````python
'''Q6.Explain the concept of generators in Python and how they are defined.

Ans:A generator in Python is a special type of iterator that allows you to iterate over a potentially large sequence of data,
but it doesn't store the entire sequence in memory at once. Instead, it generates items one at a time on-the-fly, which can be more memory-efficient,
especially for large datasets.

Generators are defined using generator functions or generator expressions.

1. Generator Functions

A generator function is a function that uses the yield keyword to produce a sequence of values,
instead of using return to return a single value. When a generator function is called, it returns an iterator that can be used to iterate over the generated values.

The yield statement allows the function to return a value and remember its state, so it can resume where it left off when next() is called again.
The function's execution is paused at each yield statement and resumed from there on the next call to next().

Defining a Generator Function:

def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # Return the current value and pause the function
        count += 1  # The function resumes here on the next iteration
Using a Generator Function:

counter = count_up_to(5)

print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
print(next(counter))  # Output: 4
print(next(counter))  # Output: 5
print(next(counter))  # Raises StopIteration, because the generator is exhausted
In this example:

When the generator function count_up_to(5) is called, it doesn't immediately execute. Instead, it returns a generator object (counter).
Each time next(counter) is called, the function executes until it reaches the yield statement, at which point it yields a value and pauses.
After all values are yielded, calling next() again raises a StopIteration exception because the generator is exhausted.

2. Generator Expressions

A generator expression is a compact way to define a generator without needing to write a full function. Itâ€™s similar to a list comprehension, but instead of producing a list, it produces a generator.

Defining a Generator Expression:

squares = (x * x for x in range(5))
In this example:

The generator expression (x * x for x in range(5)) creates a generator that produces the squares of numbers from 0 to 4.
Unlike a list comprehension, it doesnâ€™t generate and store all the values at once. It produces values lazily when iterated.
Using a Generator Expression:
python
Copy
for square in squares:
    print(square)  # Output: 0 1 4 9 16
This iteration produces values one by one without storing them in memory all at once.'''

````

````python
'''Q7. What are the advantages of using generators over regular functions?

Ans: Generators in Python provide several advantages over regular functions (which typically use `return` statements to return values). The primary benefit of using generators is their ability to handle large datasets or infinite sequences more efficiently by using **lazy evaluation**. Hereâ€™s a breakdown of the advantages of generators:

### 1. **Memory Efficiency**
- **Generators produce values one at a time**, on demand, without storing the entire sequence in memory at once. This is especially useful when dealing with large datasets or infinite sequences.
- In contrast, a regular function with a `return` statement returns all the values at once, which can lead to memory exhaustion if the data is too large.

#### Example:
Consider generating numbers from 1 to 1 million.

- **With a generator**:

  def generate_numbers():
      for i in range(1, 1000001):
          yield i
  ```
  - Here, the generator produces one number at a time, and only the current number needs to be in memory.

- **Without a generator** (using a regular function with `return`):
  def generate_numbers():
      return [i for i in range(1, 1000001)]  # All numbers are stored in memory
  ```
  - In this case, the list containing 1 million numbers is fully stored in memory at once, which could consume a lot of memory.

### 2. **Improved Performance for Large Sequences**
- Because generators generate values **on-the-fly**, they allow for **faster execution** for large sequences, as they avoid the overhead of creating and storing a large data structure like a list.
- For example, if we just need to process large data one item at a time (such as reading from a file or processing elements in a large collection), a generator is much more efficient than creating a complete list or other collection in memory.

### 3. **Lazy Evaluation (On-demand Value Generation)**
- With generators, **values are generated only when needed**, which can save both time and memory. This "lazy evaluation" means that the computation and memory usage are deferred until the value is actually requested.
- This is particularly useful when we don't need to iterate over the entire sequence, or if you're iterating over an **infinite sequence**.

#### Example:
A generator that generates an infinite sequence:
```python
def infinite_counter(start=0):
    while True:
        yield start
        start += 1

counter = infinite_counter()
for _ in range(5):
    print(next(counter))  # Output: 0 1 2 3 4
```
- The generator never creates all the numbers at once. It only produces the next value when asked, making it suitable for potentially infinite sequences.

### 4. **Convenience and Simplicity**
- Writing a generator function is **more concise and easier** than manually creating an iterator class with `__iter__()` and `__next__()` methods.
- You can create a generator function with just the `yield` keyword, which makes the code cleaner and more readable.

#### Example:
A regular function requires more effort for creating an iterator manually:
class Countdown:
    def __init__(self, start):
        self.start = start

    def __iter__(self):
        self.current = self.start
        return self

    def __next__(self):
        if self.current > 0:
            self.current -= 1
            return self.current + 1
        else:
            raise StopIteration
```
- While this code works, it's more verbose and requires additional boilerplate compared to a generator.

Using a **generator function** is much simpler:
def countdown(start):
    while start > 0:
        yield start
        start -= 1
```

### 5. **Cleaner Code in Iteration**
- Generators allow you to **iterate over large sequences without needing an explicit loop setup** or maintaining an iterator object.
- They are naturally integrated with Pythonâ€™s `for` loop, which automatically handles the `StopIteration` exception when the sequence is exhausted.

#### Example:
Using a regular function:
def get_numbers():
    return [1, 2, 3, 4, 5]

numbers = get_numbers()
for num in numbers:
    print(num)
```
- Here, the list `numbers` is stored in memory entirely before the loop runs.

With a generator:

def get_numbers():
    yield 1
    yield 2
    yield 3
    yield 4
    yield 5

for num in get_numbers():
    print(num)
```
- The generator produces numbers one by one, without storing all values in memory.

### 6. **Infinite Sequences**
- Generators are ideal for generating **infinite sequences** because they donâ€™t store all values at once. Instead, they generate values on-demand,
 which means we can have a generator that generates an infinite number of values.

#### Example:

def infinite_fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = infinite_fibonacci()
for _ in range(10):
    print(next(fib))
```
- This will produce the first 10 Fibonacci numbers, but the generator can keep producing them indefinitely without storing the entire sequence in memory.

### 7. **State Retention**
- A generator function **remembers its state** between calls to `next()`. When `yield` is encountered, the functionâ€™s state is saved, and when `next()` is called again, execution resumes from the last `yield` point.
- This is useful for implementing algorithms that require maintaining state, such as traversing large trees or graphs, or managing stateful sequences.

### **Summary of Advantages**:
| **Advantage**                     | **Generators**                                                   | **Regular Functions**                                     |
|------------------------------------|------------------------------------------------------------------|-----------------------------------------------------------|
| **Memory Efficiency**              | Generate items on-the-fly, without storing all values in memory | Store all items in memory (e.g., return lists or data structures) |
| **Performance for Large Sequences**| Faster for large datasets or infinite sequences                  | Slower for large data sets as everything is returned at once |
| **Lazy Evaluation**                | Values are produced when requested                               | All values are computed and returned at once               |
| **Simpler Code**                   | More concise and easier to write with `yield`                    | Requires more boilerplate code (e.g., implementing iterators manually) |
| **Infinite Sequences**             | Can generate infinite sequences safely                           | Cannot generate infinite sequences (would consume all memory) |

### Conclusion:
Generators are a powerful tool in Python, especially when dealing with large datasets or sequences. They are **memory-efficient**, allow for **lazy evaluation**, and

often result in **simpler, more readable code**.
Regular functions are better suited for cases where we need to return all the data at once, but for
large or infinite sequences, generators provide significant performance benefits.'''
````

````python
#4.Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(number):
    """
    Checks if a given number is prime.

    Args:
        number (int): The number to check.

    Returns:
        bool: True if the number is prime, False otherwise.
    """
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number.")
````

````python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    """
    An iterator class to generate the Fibonacci sequence up to a specified number of terms.
    """
    def __init__(self, num_terms):
        """
        Initializes the iterator.

        Args:
            num_terms (int): The number of terms in the Fibonacci sequence to generate.
        """
        self.num_terms = num_terms
        self.current_term = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        """
        Returns the iterator object itself.
        """
        return self

    def __next__(self):
        """
        Returns the next number in the Fibonacci sequence.
        """
        if self.current_term >= self.num_terms:
            raise StopIteration
        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.current_term += 1
            return self.a

# Example usage:
x=input("No of terms in the Fibonacci series")
d=int(x)
fib_iterator = FibonacciIterator(d)
for num in fib_iterator:
    print(num, end=" ")



````

````python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    """
    Yields the powers of 2 up to the given exponent.

    Args:
        max_exponent (int): The maximum exponent.

    Yields:
        int: 2 raised to the current exponent.
    """
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Example usage:
c=input("Upto power of 2 is required in the exponent?")
d=int(c)
for power in powers_of_two(d):
    print(power, end=" ")
````

````python

#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    """
    A generator function that reads a file line by line and yields each line.

    Args:
        file_path (str): The path to the file to be read.

    Yields:
        str: The next line in the file.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Strips the newline character
    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
file_path = 'example.txt'  # Replace with the path to your file
for line in read_file_line_by_line(file_path):
    print(line)


def create_text_file(file_name, content):
    """
    Creates a text file and writes the given content to it.

    Args:
        file_name (str): The name of the file to create.
        content (str): The content to write to the file.
    """
    with open(file_name, 'w') as file:
        file.write(content)

# Example usage:
file_name = "example.txt"
content = "Hello, this is a sample text file created in Python.\nThis is the second line.\nThis is the 3rd and the last line"
create_text_file(file_name, content)

````

````python
#8.Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# List of tuples
tuples_list = [(1, 3), (4, 1), (2, 5), (7, 2)]

# Sorting the list using a lambda function
sorted_list = sorted(tuples_list, key=lambda x: x[1])

# Output the result
print("Original list:", tuples_list)
print("Sorted list based on second element:", sorted_list)

````

````python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    """
    Converts a temperature from Celsius to Fahrenheit.

    Args:
        celsius (float): Temperature in Celsius.

    Returns:
        float: Temperature in Fahrenheit.
    """
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Convert to Fahrenheit using map()
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Output the results
print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
````

````python
#10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

def is_not_vowel(char):
    """
    Checks if a character is not a vowel.

    Args:
        char (str): A single character.

    Returns:
        bool: True if the character is not a vowel, False otherwise.
    """
    vowels = 'aeiouAEIOU'
    return char not in vowels

def remove_vowels(input_string):
    """
    Removes all vowels from the given string.

    Args:
        input_string (str): The input string.

    Returns:
        str: The string with vowels removed.
    """
    return ''.join(filter(is_not_vowel, input_string))

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print("Original string:", input_string)
print("String without vowels:", result)
````

````python
'''Q11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

Order Number             Book Title and Author                    Quantity                Price Per Item

   34587               Learning Python, Mark Lutz                     4                        40.95

   98762               Programming Python, Mark Lutz                  5                        56.8

   77226              Head First Python, Paul Barry                   3                        32.95

   88112            Einfuhrung in Python3, Bernd Klein                3                        24.99



Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and
the product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value
of the order is smaller than 100,00 â‚¬. Write a Python program using lambda and map.'''

#ANSWER:

# List of orders
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.8],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bernd Klein", 3, 24.99],
]

# Using map and lambda to calculate the order totals
order_totals = list(map(
    lambda order: (
        order[0],  # Order number
        order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0)  # Total with surcharge
    ),
    orders
))

# Print the result
print("Order Totals:", order_totals)







````
